### Finance

#### 毛利＆營利

[營業利益]=[銷售收入]－[銷售成本]－[銷售費用]<br>
毛利 = 銷售收入－銷售成本<br>
淨利率還包括了業外損益，所以這裡不用<br>
若同時都增加即代表：所屬產業競爭不激烈，故毛利率可以維持，公司也不用花大錢在研發新產品上，或營業費用控制得宜

```python
#做一個sort list function，不然sort(是回傳None) =>ex) [3,2,1] -> [1,2,3]
def st(given_list):
    if given_list == [None, None, None, None, None, None, None, None]:
        given_list = 'None'
    else:
        given_list.sort(reverse=True)
    return (given_list)
```

```python
gmUP = []
omUP = []
for sid in sids:
    if tw0050[sid]['Finance']['營業毛利率']['Data'] == st(tw0050[sid]['Finance']['營業毛利率']['Data']):
        gmUP.append(sid)
    if tw0050[sid]['Finance']['營業利益率']['Data'] == st(tw0050[sid]['Finance']['營業利益率']['Data']):
        omUP.append(sid)
    
growGMOM = list(set(gmUP) & set(omUP))
print(growGMOM)
```

```python
#找出最高毛利之企業
maxGM = 0
for sid in growGMOM:
    if tw0050[sid]['Finance']['營業毛利率']['Data'][0] > maxGM:
        maxGM = tw0050[sid]['Finance']['營業毛利率']['Data'][0]
        maxsid =sid
print(maxsid)
```

```python
roaUP = []
for sid in sids:
    temp = []
    for i in tw0050[sid]['Finance']['資產報酬率']['Data']:
        if i == None:
            pass
        else:
            temp.append(i)
    if temp == []:
        pass
    else:
        for j in range(len(temp[:-1])):
            if (temp[j] - temp[j+1]) > (-0.5):
                test = temp[j+1]
            else:
                break
        if test == temp[-1]:
            roaUP.append(sid)
roaUP
```

```python
roeUP = []
for sid in sids:
    temp = []
    for i in tw0050[sid]['Finance']['股東權益報酬率']['Data']:
        if i == None:
            pass
        else:
            temp.append(i)
    if temp == []:
        pass
    else:
        for j in range(len(temp[:-1])):
            if (temp[j] - temp[j+1]) > (-0.5):
                test = temp[j+1]
            else:
                break
        if test == temp[-1]:
            roeUP.append(sid)
roeUP
```

#### 流動比率 & 速動比率
流動比率(Current Ratio) -> 包含存貨 > 200%<br>
速動比率(Quick Ratio) -> 去掉存貨 > 100%

```python
goodCurrent = []
goodQuick = []
for sid in sids:
    if tw0050[sid]['Finance']['速動比率']['Data'][0] != None:
        if tw0050[sid]['Finance']['速動比率']['Data'][0] > 100:
            goodCurrent.append(sid)
    if tw0050[sid]['Finance']['流動比率']['Data'][0] != None:
        if tw0050[sid]['Finance']['流動比率']['Data'][0] > 200:
            goodQuick.append(sid)   
print(goodCurrent)
print(goodQuick)
```

```python
goodCurrentQuick = list(set(goodCurrent) & set(goodQuick))
print(goodCurrentQuick)
```

```python
goodCurrentQuickDF = pd.DataFrame([{'Sid':'1326',\
                                   'Current':tw0050['1326']['Finance']['流動比率']['Data'][0],\
                                'Quick':tw0050['1326']['Finance']['速動比率']['Data'][0]}])
goodCurrentQuickDF.set_index('Sid',inplace=True)

goodCurrentQuickDF
```

```python
for sid in goodCurrentQuick[1:]:
    temp = pd.DataFrame([{'Sid':sid,\
                                   'Current':tw0050[sid]['Finance']['流動比率']['Data'][0],\
                                'Quick':tw0050[sid]['Finance']['速動比率']['Data'][0]}])
    temp.set_index('Sid',inplace=True)
    goodCurrentQuickDF = pd.concat([goodCurrentQuickDF,temp])
goodCurrentQuickDF
```

```python
goodCurrentQuickDF.sort_values(by='Current',ascending=False).plot.bar(figsize=(15,6))
plt.title('Current Ratio(>200%) & Quick Ratio(>100%)',fontsize=20)
plt.ylabel('Current & Quick Ratio',fontsize=20)
plt.xlabel('')
plt.tick_params(axis='x', which='major', labelsize=14)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.legend(prop={'size': 13})
#plt.savefig('Current_Quick.jpg',bbox_inches='tight')
```
